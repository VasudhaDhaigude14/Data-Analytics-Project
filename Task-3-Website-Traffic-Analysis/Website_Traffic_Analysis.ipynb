```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("C:\\Users\\HP\\Downloads\\archive (2)\\traffic.csv", encoding="latin1")

print(df.head())
print(df.columns.tolist())
```

```python
df.drop_duplicates(inplace=True)

df.fillna(method='ffill', inplace=True)

# Convert date column (change column name if needed)
df['Date'] = pd.to_datetime(df['date'], errors='coerce')

print(df.isnull().sum())
```

```python
print("Total Events :", len(df))

print("Unique Links :", df["linkid"].nunique())

print("Unique Tracks :", df["track"].nunique())

print("Unique Artists :", df["artist"].nunique())

print("Countries :", df["country"].nunique())
```

```python
top_links = df["linkid"].value_counts().head(10)

plt.figure(figsize=(10,5))
top_links.plot(kind="bar")

plt.title("Top 10 Most Visited Links")
plt.ylabel("Visits")
plt.xlabel("Link ID")
plt.xticks(rotation=45)

plt.show()
```

```python
top_country = df["country"].value_counts().head(10)

plt.figure(figsize=(10,5))
top_country.plot(kind="bar", color="orange")

plt.title("Top Countries")
plt.ylabel("Visitors")

plt.show()
```

```python
top_artist = df["artist"].value_counts().head(10)

plt.figure(figsize=(10,5))
top_artist.plot(kind="bar", color="green")

plt.title("Top Artists")

plt.ylabel("Clicks")

plt.xticks(rotation=45)

plt.show()
```

```python
daily = df.groupby("date").size()

plt.figure(figsize=(12,5))

daily.plot(marker="o")

plt.title("Daily Website Traffic")

plt.ylabel("Events")

plt.show()
```

```python
event = df["event"].value_counts()

plt.figure(figsize=(6,6))

event.plot(kind="pie", autopct="%1.1f%%")

plt.ylabel("")

plt.title("Event Distribution")

plt.show()
```
